```python
class Human:
    # define properties of the class object "Human"
    def __init__(self,n,o):
        self.name = n
        self.occupation = o
    # define mthods of the class object "Human"
    def do_work(self):
        if self.occupation == "Actor":
            print(self.name, "plays tennis")
        elif self.occupation == "Tennis Player":
            print(self.name,"Shoots Movie")
    def speak(self):
        if self.name == "Tom Cruise":
            print(self.name, "says he was Ethan Hunt of Mission Impossible")
        elif self.name == "Maria Sharapova":
            print(self.name,"says she won 5 grand slams")
```

```python
# create an instance for the class "Human" and assign it to a variable object
tom = Human("Tom Cruise","Actor")
```

```python
# Let's call the defined methods for the class "Human"
tom.do_work()
tom.speak()
```

```python
# create another instance of the class "Human"
maria = Human("Maria Sharapova","Tennis Player")
```

```python
# Let's call the defined methods for the class "Human"
maria.do_work()
maria.speak()
```
